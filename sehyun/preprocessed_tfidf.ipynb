```python
import pandas as pd
```

```python
npp_text = pd.read_csv('data(preprocessing_okt).csv')
tmp_text = pd.read_csv('tmp(preprocessing_okt).csv')
```

결측치를 제거하였습니다.


```python
npp_text = npp_text.token
tmp_text = tmp_text.token
```

```python
# stop_words 지정 : 개인적으로 지정

stop_words = ['하다','있다','되다','없다','않다','아니다','이다','받다','되어다','보다','이렇다','그렇다']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer(max_features = 10000,stop_words = stop_words)

# max_features 를 조정할 때 마다 단어가 좀 달라지는 문제가 있습니다. 

# matrix 생성
npp_matrix = tfidf_vectorizer.fit_transform(npp_text)
tmp_matrix = tfidf_vectorizer.fit_transform(tmp_text)


feature_names = tfidf_vectorizer.get_feature_names_out()

npp_scores = npp_matrix.toarray()
tmp_scores = tmp_matrix.toarray()

```

```python
# topword 추출

npp_topwords_indices = npp_scores.mean(axis=0).argsort()[::-1][:100]
tmp_topwords_indices = tmp_scores.mean(axis=0).argsort()[::-1][:100]

npp_top_words = [(feature_names[idx], npp_scores.mean(axis=0)[idx]) for idx in npp_topwords_indices]
tmp_top_words = [(feature_names[idx], tmp_scores.mean(axis=0)[idx]) for idx in tmp_topwords_indices]


print("NPP'S Top 100 Words with High TF-IDF Scores:")
for word, score in npp_top_words:
    print(f"{word}: {score:.4f}")

print("\nTMP'S Top 100 Words with High TF-IDF Scores:")
for word, score in tmp_top_words:
    print(f"{word}: {score:.4f}")
```

```python
word_differences = [(feature_names[idx], npp_scores.mean(axis=0)[idx] - tmp_scores.mean(axis=0)[idx]) for idx in npp_topwords_indices]

sorted_word_differences = sorted(word_differences, key=lambda x: x[1], reverse=True)
print("Top Words with Largest TF-IDF Difference (NPP - TMP):")
for word, difference in sorted_word_differences:
    print(f"{word}: {difference:.4f}")
```

### max_features 를 조정할 때 마다 단어가 좀 달라지는 문제가 있습니다. 비교를 한번 해보고자 합니다.

```python
npp_text = pd.read_csv('data(preprocessing_okt).csv')
tmp_text = pd.read_csv('tmp(preprocessing_okt).csv')

tmp_text = tmp_text.dropna()

npp_text = npp_text.token
tmp_text = tmp_text.token

stop_words = ['하다','있다','되다','없다','않다','아니다','이다','받다','되어다','보다','이렇다','그렇다']
```

```python
# 10000은 빼고 하려고 합니다. (너무 지엽적인 단어들이 들어와서)

max_features_values = [100, 500, 1000, 5000]

for max_features in max_features_values:
    print(f'with max_features :  {max_features}')
    tfidf_vectorizer = TfidfVectorizer(max_features=max_features, stop_words=stop_words)
    npp_matrix = tfidf_vectorizer.fit_transform(npp_text)
    tmp_matrix = tfidf_vectorizer.transform(tmp_text)
    feature_names = tfidf_vectorizer.get_feature_names_out()

    npp_scores = npp_matrix.toarray()
    tmp_scores = tmp_matrix.toarray()
    
    npp_topwords_indices = npp_scores.mean(axis=0).argsort()[::-1][:100]
    tmp_topwords_indices = tmp_scores.mean(axis=0).argsort()[::-1][:100]
    
    npp_top_words = [(feature_names[idx], npp_scores.mean(axis=0)[idx]) for idx in npp_topwords_indices]
    tmp_top_words = [(feature_names[idx], tmp_scores.mean(axis=0)[idx]) for idx in tmp_topwords_indices]
    
    print("NPP'S Top 100 Words with High TF-IDF Scores:")
    for word, score in npp_top_words:
        print(f"{word}: {score:.4f}")

    print("\nTMP'S Top 100 Words with High TF-IDF Scores:")
    for word, score in tmp_top_words:
        print(f"{word}: {score:.4f}")
    
    
    word_differences = [(feature_names[idx], npp_scores.mean(axis=0)[idx] - tmp_scores.mean(axis=0)[idx]) for idx in npp_topwords_indices]

    # npp tfidf 점수가 높은 순
    npp_word_differences = sorted(word_differences, key=lambda x: x[1], reverse=True)
    print(f"Top Words with Largest TF-IDF Difference NPP word")
    
    for word, difference in npp_word_differences:
        print(f"{word}: {difference:.4f}")
        
    # tmp tfidf 점수가 높은 순
    tmp_word_differences = sorted(word_differences, key=lambda x: x[1], reverse=False)
    print(f"Top Words with Largest TF-IDF Difference TMP word")
    
    for word, difference in tmp_word_differences:
        print(f"{word}: {difference:.4f}")
        
    print("\n")
    
```
