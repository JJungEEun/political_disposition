```python
# import 

import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from konlpy.tag import Okt
```

```python
# data load

data = pd.read_csv("NPP_final_0729.csv")
```

```python
data.head()
```

```python
# texts 만 따로 빼서 전처리 함
# 일단 500개만

texts = data.content
# texts = data.content[:500]
texts.shape
```

```python
# Okt를 토크나이저로..

tokenizer = Okt()
```

```python
# 전처리 드가자

preprocessed_texts = []
for text in texts:
    tokens = tokenizer.pos(text, norm=True, stem = True)
    preprocessed_text = " ".join([word for word, pos in tokens if pos in ['Noun','Verb','Adjective']])
    preprocessed_texts.append([preprocessed_text])

preprocessed_texts[0]
```

```python
# 정은님이 주신 거

npp = pd.read_csv('data(preprocessing_okt).csv')
texts = npp.token
texts.head(10)
```

```python
# 결과 확인

preprocessed_texts[1:3]
```

```python
# list of list -> list of strings 로 변경.
texts = [" ".join(sublist) for sublist in preprocessed_texts]
print(texts[1:3])
```

```python
# stop_word 따로 지정

stop_words = ['하다','있다','되다','없다','않다','아니다','이다','받다','되어다','보다','이렇다','그렇다']
```

```python
# tfidfVectorizer 생성 및 적용



tfidf_vectorizer = TfidfVectorizer(max_features = 100,stop_words = stop_words)
tfidf_matrix = tfidf_vectorizer.fit_transform(texts)
feature_names = tfidf_vectorizer.get_feature_names_out()
tfidf_scores = tfidf_matrix.toarray()

```

```python
# top word 100개 추출

top_words_indices = tfidf_scores.mean(axis=0).argsort()[::-1][:100]  # Sort indices based on mean TF-IDF score
top_words = [feature_names[idx] for idx in top_words_indices]

print("Top 100 Words with High TF-IDF Scores:")
print(", ".join(top_words))
```

```python
#문서별 키워드 3개 추출 -> topic?

keywords = []
for i, text in enumerate(texts):
    tfidf_scores_text = tfidf_scores[i]
    sorted_indices = tfidf_scores_text.argsort()[::-1]
    top_keywords = [feature_names[idx] for idx in sorted_indices[:10]]
    keywords.append(top_keywords)
    
for i,text_keywords in enumerate(keywords):
    print(f'Keyword {i+1} : {text_keywords}')
```
